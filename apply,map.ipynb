```python
date_list = [{'yyyy-mm-dd': '2000-06-27'},
         {'yyyy-mm-dd': '2002-09-24'},
         {'yyyy-mm-dd': '2005-12-20'}]
df = pd.DataFrame(date_list, columns = ['yyyy-mm-dd'])
df
```

```python
def ex_year(row):
  return row.split('-')[0]
```

```python
df['year']=df['yyyy-mm-dd'].apply(ex_year)
df
```

```python
def ex_age(year,current):
  return current-int(year) 
```

```python
df['age']=df['year'].apply(ex_age,current=2022)
df
```

```python
def intro(age,pre,suf):
  return pre+str(age)+suf
```

```python
df['intro']=df['age'].apply(intro,pre='i am',suf='years old')
df
```
